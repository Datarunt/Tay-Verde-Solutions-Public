### Remove unecessary characters  - All set

```python
cleaned_parts = []

for line in new_df["Material"]:
    if pd.isnull(line):
        cleaned_parts.append(None)
        continue

    line = str(line)
    tokens = re.findall(r'\w[\w-]*', line)

    cleaned_tokens = []
    for tok in tokens:
        # Remove trailing letters (one or more) at the end of the token
        cleaned_tok = re.sub(r'[A-Za-z]+$', '', tok)
        # Skip if cleaned token is empty or only letters
        if cleaned_tok and not re.fullmatch(r'[A-Za-z]+', cleaned_tok):
            cleaned_tokens.append(cleaned_tok)

    # Take the first cleaned token if exists, else None
    cleaned_parts.append(cleaned_tokens[0] if cleaned_tokens else None)

print(f"Lengths: DataFrame={len(new_df)}, cleaned_parts={len(cleaned_parts)}")

if len(cleaned_parts) == len(new_df):
    new_df["Material"] = cleaned_parts
else:
    raise ValueError("Length mismatch between cleaned parts and DataFrame rows!")
```

### Adjust Data Formatting - All set

```python
new_df['Date Due'] = pd.to_datetime(new_df['Date Due'])
new_df['Date Rcvd'] = pd.to_datetime(new_df['Date Rcvd'])
new_df['Days Late'] = pd.to_numeric(new_df['Days Late'], errors='coerce')

# Vendor Encoding
vendor_perf = new_df.groupby('Vendor Name')['Days Late'].mean().reset_index()
vendor_perf.columns = ['Vendor Name', 'Avg Days Late']

# Merge back to original
new_df = new_df.merge(vendor_perf, on='Vendor Name')
```

### Remove duplicates - All set

```python
# === BEFORE removing duplicates ===
# Find duplicated rows (across all columns)
duplicates_before = df_missing_cleaned[df_missing_cleaned.duplicated(keep=False)]

# Count unique duplicate groups
num_unique_duplicate_groups_before = duplicates_before.groupby(list(df_missing_cleaned.columns)).ngroups

# Print duplicate info before
print("Before removing duplicates:")
print(f"Total duplicate rows found: {len(duplicates_before)}")
print(f"Unique duplicate groups: {num_unique_duplicate_groups_before}")
print(f"Number of rows before dropping duplicates: {len(df_missing_cleaned)}")

# === REMOVE duplicates ===
df_no_duplicates = df_missing_cleaned.drop_duplicates()

# === AFTER removing duplicates ===
# Find duplicates again to confirm they're removed
duplicates_after = df_no_duplicates[df_no_duplicates.duplicated(keep=False)]
num_unique_duplicate_groups_after = duplicates_after.groupby(list(df_no_duplicates.columns)).ngroups if not duplicates_after.empty else 0

# Print duplicate info after
print("\nAfter removing duplicates:")
print(f"Total duplicate rows found: {len(duplicates_after)}")
print(f"Unique duplicate groups: {num_unique_duplicate_groups_after}")
print(f"Number of rows after dropping duplicates: {len(df_no_duplicates)}")

Final_df = df_no_duplicates.dropna(subset=['Days Late'])
```

# Control chart - All set

```python
# Resample by week and compute average Days Late
weekly_late = Final_df.resample('W', on='Date Rcvd')['Days Late'].mean().dropna()

# Step 3: Calculate control limits
mean = weekly_late.mean()
std = weekly_late.std()
ucl = mean + 3 * std
lcl = mean - 3 * std

# Step 4: Plot control chart
plt.figure(figsize=(14, 6))
plt.plot(weekly_late.index, weekly_late, marker='o', label='Avg Days Late')
plt.axhline(mean, color='green', linestyle='--', label='Mean')
plt.axhline(ucl, color='red', linestyle='--', label='UCL (+3σ)')
plt.axhline(lcl, color='red', linestyle='--', label='LCL (−3σ)')

# Highlight points outside control limits
out_of_control = weekly_late[(weekly_late > ucl) | (weekly_late < lcl)]
plt.scatter(out_of_control.index, out_of_control, color='red', zorder=5, label='Out of Control')

plt.title('Control Chart: Average Days Late by Week')
plt.xlabel('Week')
plt.ylabel('Average Days Late')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Hypothesis Testing ( ANOVA) - All Set

```python
# 2. Fix dtypes
Final_df['Days Late'] = pd.to_numeric(Final_df['Days Late'], errors='coerce')
Final_df['Date Rcvd'] = pd.to_datetime(Final_df['Date Rcvd'], errors='coerce')
Final_df['Date Due'] = pd.to_datetime(Final_df['Date Due'], errors='coerce')

# 3. Drop missing values needed for ANOVA
anova_df = Final_df.dropna(subset=[
    'Qty Rcvd', 'Days Late', 'Net Order Price',
    'Vendor Name', 'Material', 'Description',
    'Status (Qty)', 'Status (Date)', 'Military Contract No.',
    'Description p. group', 'Comm Code Desc'
])

# 4. Define safe ANOVA runner using Q() for column names
def run_anova(df, dependent_var, factor_var):
    try:
        # Q() safely wraps variable names with spaces
        formula = f'Q("{dependent_var}") ~ C(Q("{factor_var}"))'
        model = ols(formula, data=df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)

        print(f"\n🔍 ANOVA: {dependent_var} ~ {factor_var}")
        print(anova_table)

        p_value = anova_table["PR(>F)"].iloc[0]
        if p_value < 0.05:
            print("✅ Significant effect (p < 0.05)")
        else:
            print("⚠️ No significant effect (p >= 0.05)")

    except Exception as e:
        print(f"❌ Error in ANOVA for {dependent_var} ~ {factor_var}: {e}")

# 5. Define dependent and categorical variables
dependent_vars = ['Qty Rcvd', 'Days Late', 'Net Order Price']
categorical_factors = [
    'Vendor Name', 'Material', 'Description', 'Status (Qty)',
    'Status (Date)', 'Military Contract No.',
    'Description p. group', 'Comm Code Desc'
]

# 6. Run all ANOVA tests
for dep_var in dependent_vars:
    for factor in categorical_factors:
        run_anova(anova_df, dep_var, factor)
```
